```python
import statsmodels.api as sm
import numpy as np
import statsmodels.api as sm
import time
import pandas as pd
```

```python
# Simulate data
np.random.seed(123)
n = 100000
x = np.random.randn(n)
y = 1 + 2 * x + np.random.randn(n)

# Add intercept
X = sm.add_constant(x)

# Fit quantile regression at median (q = 0.5)
start_time = time.time()
model = sm.QuantReg(y, X).fit(q=0.1, max_iter = 1000)
end_time = time.time()

# Summary and timing
print(model.summary())
print("Time taken (Python):", end_time - start_time)
```

```python
quantiles = np.round(np.linspace(0.1,0.9,9), decimals=2)
df = df.dropna()

y_train = df["target"]
start_time = time.time()
for q in quantiles:
    x_train = sm.add_constant(df[str(q)])
    model = sm.QuantReg(y_train, x_train).fit(q=q)
end_time = time.time()
print("Time taken (Python):", end_time - start_time)
```
